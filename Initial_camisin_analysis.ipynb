```python
import pandas as pd
from functools import reduce
import glob 
```

```python
df_metadata = pd.read_table('C:\Livia\Doutorado\Tese_escrita\Analysis\pre_metadata.tsv',header=None,names=['assembly_accession','tax_id','organism_name'])
df_metadata
```

# 30 genomes

```python
df_selected_30 = pd.read_table('C:\Livia\Doutorado\Tese_escrita\Analysis\Cami_30g_6s\genomes_selecionados.txt',header=None, names=['assembly_accession'])
df_selected_30.head()
```

```python
df_merged = pd.merge(df_metadata,df_selected_30)
df_merged
```

```python
dataframe_list = list()
for name in glob.glob('Cami_30g_6s/distributions/' + 'distr*.txt'):
    df_temp = pd.read_table(name, header=None, names=['assembly_accession','frequency'])
    dataframe_list.append(df_temp)

df_distr = reduce(lambda left,right: pd.merge(left,right,on='assembly_accession'), dataframe_list)
df_distr.columns = ['assembly_accession','distribution_0','distribution_1','distribution_2','distribution_3','distribution_4','distribution_5']
df_distr
```

```python
df_final = pd.merge(df_merged,df_distr)
df_final.loc[:,'distribution_0':] = df_final.loc[:,'distribution_0':].astype(str).replace('\.',',', regex=True)
df_final
```

```python
df_final.to_csv('Cami_30g_6s/selected_genomes_complete_30.tsv', sep ='\t', index=None)
```
