# Spades careful analysis

```python
#path="Cami_30g_6s/S_careful/"
#path="Cami_100g_6s/S_careful/"
path="Cami_200g_6s/S_careful/"


df_checkm = pd.read_table(path+'ext_summary_mod.txt', header=None, sep='\s{2,}', engine='python')
column = ['Bin_Id','Marker_lineage','n_genomes','n_markers','n_marker_sets','Completeness','Contamination','Strain_heterogeneity','Genome_size','ambiguous_bases','n_scaffolds','n_contigs','N50_scaf','N50_contigs','Mean_scaffold_length','Mean_contig_length','Longest_scaffold','Longest_contig','GC','GC_std','Coding_density','Translation_table','n_predicted_genes','0','1','2','3','4','5+']
df_checkm.columns = column
df_checkm = df_checkm[~df_checkm['Bin_Id'].str.contains('sample')]

recovered_bins_initial = df_checkm.shape[0]

all = len(df_checkm[df_checkm['Bin_Id'].str.contains('_all_bin')])
meta = len(df_checkm[df_checkm['Bin_Id'].str.contains(r'(?=^MAG)(?=.*meta)',regex=True)])
careful = len(df_checkm[df_checkm['Bin_Id'].str.contains(r'(?=^MAG)(?=.*care)',regex=True)])

df_checkm = df_checkm.loc[df_checkm['Completeness'] >= 70.0 ]
df_checkm = df_checkm.set_index('Bin_Id')
recovered_bins_final = df_checkm.shape[0]

all_f = len(df_checkm[df_checkm.index.str.contains('_all_bin')])
meta_f = len(df_checkm[df_checkm.index.str.contains(r'(?=^MAG)(?=.*meta)',regex=True)])
careful_f = len(df_checkm[df_checkm.index.str.contains(r'(?=^MAG)(?=.*care)',regex=True)])

print(f'Initial number of bins {recovered_bins_initial} - all = {all} | meta = {meta} | careful = {careful} \nnumber of bins greater than 70% {recovered_bins_final} -- all = {all_f} | meta = {meta_f} | careful = {careful_f}')

df_checkm

```

```python
lista_df=list()

file = open(path+"MAGs_2_analise.txt", "w")  

with open (path+"drep_report_MAGs_cami.txt",'r+') as f:
    count = 0 
    for i,line in enumerate(f):
        same_mag = line.replace('.fa','').strip().split(";")
        if any("_genomic" in item for item in same_mag) and any("MAG" in item for item in same_mag):
            
            ref=""            
            df_same = df_checkm[df_checkm.index.isin(same_mag)].copy()
            
            for value in same_mag:
                if '_genomic' in value:                    
                    ref+= value+' '
                if any(x in value for x in ['all','MAG']):
                    count +=1                    
                    file.write(f'{value}\n') 
            df_same['Reference'] = ref
            #opa.append(ref)
            lista_df.append(df_same)
            
f.close()
file.close()

'''
s = set(temp2)
temp3 = [x for x in temp1 if x not in s]"
'''

final_df = pd.concat(lista_df, ignore_index=False)
final_df = final_df[['Reference','Completeness','Contamination','Strain_heterogeneity','n_contigs','Longest_contig']].reset_index()
final_df = final_df[final_df['Bin_Id'].str.contains('MAG|all',regex=True)]
#final_df.groupby('Reference').size()
#final_df = final_df[~final_df['Reference'].str.contains('GCF_001761325.1_ASM176132v1_genomic')]
final_df.head()
```

```python
ckm = df_checkm.reset_index()[['Bin_Id']]

ref_not_found = pd.merge(ckm, final_df[['Bin_Id','Reference']], how = 'outer')
ref_not_found = ref_not_found[ref_not_found['Reference'].isnull()]
ref_not_found.count()
```

```python
## Analyzing the reference genomes that were found only by co-assembly approach
df_coa = pd.read_table(path+"drep_report_MAGs_cami.txt", header = None)
df_coa[0] = df_coa[0].str.replace('.fa','')
df_coa_notdetect = df_coa.loc[(~df_coa[0].str.contains('MAG')) & (~df_coa[0].str.contains('all'))]
df_coa_notdetect = df_coa_notdetect[0].str.split(';', expand=True)

number_bins_only_coa = 0 
for i in range(df_coa_notdetect.shape[1]):
    number_bins_only_coa += len(df_coa_notdetect[df_coa_notdetect[i].str.contains('GCF',na=False)]) 

print(f'{number_bins_only_coa} - reference genomes not found')

df_coa_all = df_coa.loc[(~df_coa[0].str.contains('MAG'))]
df_coa_all = df_coa_all[0].str.split(';', expand=True)
 
number_bins_only_coa = 0 
for i in range(1,df_coa_all.shape[1]):
    number_bins_only_coa += len(df_coa_all[df_coa_all[i].str.contains('all_bin',na=False)]) 

print(f'{number_bins_only_coa} - co-assembly bins with ref genomes similarity')

df_coa_mag = df_coa.loc[~df_coa[0].str.contains('_all')]
df_coa_mag = df_coa_mag[0].str.split(';', expand=True)
 
number_bins_only_coa = 0 
for i in range(1,df_coa_mag.shape[1]):
    number_bins_only_coa += len(df_coa_mag[df_coa_mag[i].str.contains('MAG',na=False)]) 

print(f'{number_bins_only_coa} - screening bins with ref genomes similarity')
```

```python
df_coa_all_match = df_coa.loc[(df_coa[0].str.contains('GCF')) & (df_coa[0].str.contains('all'))]
df_coa_all_match = df_coa_all_match[0].str.split(';', expand=True)
print(df_coa_all_match.shape, 'all')

df_coa_meta_match = df_coa.loc[(df_coa[0].str.contains('GCF')) & (df_coa[0].str.contains('MAG')) & (df_coa[0].str.contains('meta'))]
df_coa_meta_match = df_coa_meta_match[0].str.split(';', expand=True)
print(df_coa_meta_match.shape, 'meta')

df_coa_care_match = df_coa.loc[(df_coa[0].str.contains('GCF')) & (df_coa[0].str.contains('MAG')) & (df_coa[0].str.contains('care'))]
df_coa_care_match = df_coa_care_match[0].str.split(';', expand=True)
print(df_coa_care_match.shape, 'care')
```

```python
df_error = pd.read_table(path+'error_detection.txt', header=None, names=['pilon_errors','Bin_Id'])
df_error.head()
```

```python
final_df = pd.merge(final_df,df_error, on='Bin_Id')
final_df.head()
```

```python
conditions = [
    (final_df['Bin_Id'].str.contains('MAG.*care',regex=True) ),
    (final_df['Bin_Id'].str.contains('MAG.*meta',regex=True) ),
    (final_df['Bin_Id'].str.contains('all',regex=True) )
    ]

# create a list of the values we want to assign for each condition
values = ['careful', 'meta', 'all']

# create a new column and use np.select to assign values to it using our lists as arguments
final_df['Category'] = np.select(conditions, values)
final_df['Assembly'] = 'Spades_careful'
final_scare = final_df.copy()
final_scare.to_csv(path+'Final_file_organized.tsv', sep='\t', index=False)
final_scare.head()
```
